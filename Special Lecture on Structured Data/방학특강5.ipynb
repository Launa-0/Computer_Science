```python
import folium

folium.Map()
```

```python
map=folium.Map(location=[37.6071,127.0424],
               zoom_start=15,
               zoom_control=True,
               control_scale=True)
map
```

```python
mk=folium.Marker([37.6071,127.0424],popup=folium.Popup('동덕여자대학교',max_width=1000))
mk.add_to(map)

map
```

```python
mk2=folium.Marker([37.6075,127.042],tooltip='위치2')
mk2.add_to(map)

map
```

```python
mk3=folium.Marker([37.607,127.043],tooltip='위치3',icon=folium.Icon())
mk3.add_to(map)

map
```

```python
mk4=folium.Marker([37.607,127.042],tooltip='위치4',icon=folium.Icon(color='red',icons='hippo',prefix='fa'))
mk4.add_to(map)

map
```

```python
mk5=folium.Circle([37.606,127.042],tooltip='위치5',color='green')
mk5.add_to(map)

map
```

```python
mk6=folium.Circle([37.606,127.042],tooltip='위치6',color='green',radius=100)
mk6.add_to(map)

map
```

```python
mk7=folium.Circle([37.605,127.041],tooltip='위치7',color='green',radius=300,fill=True)
mk7.add_to(map)

map
```
